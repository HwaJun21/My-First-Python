```python
import numpy as np


# 브로드캐스팅을 사용X
x = np.array([1,2,3,4,5])
y = np.array([1,1,1,1,1])
print(x+y)

# 브로드캐스팅을 사용한 경우
y = 1
print(x+y)
```

브로드캐스팅 연산
  - shape가 다른 두 배열의 연산
    1. shape가 다른 두 배열 사이의 이항연산에서 브로드캐스팅이 발생
    2. 두 배열을 같은 shape로 만든 후 연산 수행

```python
# shape가 다른 두 배열의 연산
a = np.arange(1,25).reshape(4,6)
print(a*100)   # 브로드캐스팅

print("------과정------")
print("--1. 스칼라의 배열 변경--")
new = np.full_like(a, 100)
print(new)
print("--2. 배열 이항 연산--")
print(a*new)
```

```python
# shape가 다른 배열 연산
a = np.arange(5).reshape(1,5)
print(a)
b = np.arange(5).reshape(5,1)
print(b)
print(a+b)
```

배열요소 추가및 삭제
  - np.resize(a, new)
  - np.ndarray.resize(new, refcheck = True)
  - 배열의 shape와 크기를 변경

```python
# resize() : 요소가 동일한 경우 reshape()와 같다
a = np.arange(20)
y = np.resize(a, (5,4))
print(y)
y
```

```python
# resize() : 요소가 더 많은 경우 -> recursive 된다
a = np.arange(20)
y = np.resize(a, (5,6))
print(y)
y
```

```python
# resize와 reshape의 차이
x = np.arange(20)
y = x.reshape(4,5)
z = np.resize(x, (4,5))

print(x)
print(y)
print(z)

x[0] = 10
print(x)
print(y)
print(z)

y[0,0] = 100
print(x)
print(y)
print(z)
```

배열 요소 추가및 삭제
  - np.append(arr, values, axis=None)
  - 배열의 끝에 값을 추가

```python
# 값 추가
a = np.arange(3)
print(a)
b = np.append(a, 3)
print(b)
```

```python
# 배열추가
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[7,8,9]])
c = np.append(a, b)
print(c)
```

```python
# 축(axis) 추가
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[7,8,9]])
c = np.append(a, b, axis = 0)
print(c)
```

```python
# 축(axis) 추가
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[7,8,9]])
c = np.append(a, b, axis = 1)
print(c)
```

  - np.insert(arr, obj, values, axis=None)
    1. axis를 지정하지 않으면 1차원 배열로 변환
    2. 추가할 방향을 axis로 지정

```python
# 값 추가 : 축지정
a = np.array([[1,1],[2,2],[3,3]])
b = np.insert(a, 1, 5, axis = 0)
c = np.insert(a, 1, 5, axis = 1)
print(a)
print(b)
print(c)
```

```python
# 값 추가 : 1차원 배열
a = np.array([1,2,3])
b = np.insert(a, 1, 5)
print(a)
print(b)
b = np.insert(b, 2, 6)
print(b)
```

배열 요소 삭제
  - np.delete(arr, obj, axis=None)
    1. axis 지정하지 않으면 1차원 배열로 변환
    2. 삭제할 방향을 axis로 지정
    3. delete함수는 원본 배열을 변경하지 않으면 새로운 배열을 반환

```python
# 값 삭제
a = np.arange(12).reshape(3,4)
print(a)
print(np.delete(a, 5))
```

```python
# 값 삭제 : 축지정
a = np.arange(12).reshape(3,4)
print(a)
print(np.delete(a, 1, axis = 0))
print(np.delete(a, 1, axis = 1))
```

배열 결합

```python
# concatenate()
a1 = np.array([1,2,3])
a2 = np.array([4,5,6])
arr = np.concatenate((a1, a2))
print(arr)
```

```python
# concatenate() : 축지정
a1 = np.array([[1,2,3]])
a2 = np.array([[4,5,6]])
arr = np.concatenate((a1, a2), axis = 0)
arr1 = np.concatenate((a1, a2), axis = 1)
print(arr)
print(arr1)
```

 - np.vstack(튜플)
  1. 튜플로 설정된 여러 배열을 수직방향으로 연결(axis=0)
  2. np.concatenate(tup, axis=0)

 - np.hstack(튜플)
  1. 튜플로 설정된 여러 배열을 수평방향으로 연결(axis=1)
  2. np.concatenate(tup, axis=1)

```python
a1 = np.array([[1,2,3]])
a2 = np.array([[4,5,6]])

print(np.vstack((a1, a2)))
print(np.concatenate((a1, a2), axis = 0))
```

```python
a1 = np.array([[1,2,3]])
a2 = np.array([[4,5,6]])

print(np.hstack((a1, a2)))
print(np.concatenate((a1, a2), axis = 1))
```
