# 문자열 재정렬
# 1. 입력받은 문자를 오름차순 정렬하여 출력
# 2. 숫자의 경우 더한 합계를  출력
# 3. 문자열 입력 : K1KA5CB7
# 4. 재정렬된 출력 : ABCKK13

```python
data = input("문자열 입력 : ")
res = []
value = 0

for i in data:
  if i.isalpha():
    res.append(i)
  else:
    value += int(i)

res.sort()

if value != 0:
  res.append(str(value))

print(''.join(res))

```

리스트 컴프리헨션
  - for문과 if문이 있는 구조에서 리스트를 생성해주는 문법
  - 단, 1차원 이상은 권장하지 않는다
  - [표현식 for 아이템 in 이터러블]
  - [표현식 for 아이템 in 이터러블 단순조건문]
  - [표현식 조건문 else 표현식 for 아이템 in 이터러블]


```python
# 2의 배수 출력

even = []
for i in range(11):
  if i % 2 == 0:
    even.append(i)

print(even)
```

```python
# 리스트 컴프리헨션

even = [i for i in range(11) if i % 2 == 0]
print(even)
```

튜플 수정
  - 튜플을 만들어 놓으면 요소의 추가, 수정, 삭제 불가
  - packing, unpacking
    1. packing : 한 변수에 여러개의 데이터를 할당하여 여러개의 객체를 하나의 객체로 합쳐주는 것 -> 묶는다 -> 리스트나 튜플을 이용해서 선언
    2. unpacking : 한 변쉬의 데이터를 각각의 변수로 반환하는 것 -> 묶은걸 푼다

```python
f = ('사과','포도','수박','참외','배')
print(f)
```

```python
# unpacking
fruit1, fruit2, fruit3, fruit4, fruit5 = f
print(fruit1)
print(fruit4)
```

튜플 구조를 만들어주는 enumertae()
  - 인덱스와 원소를 동시에 접근하면서 반복문을 반복 할 때 사용
  - 기본적으로 인덱스와 원소로 이루어진 튜플을 만들어 준다
  - 튜플 packing구조로 만들어 준다
  - 사용
    1. for in enumerate():

```python
# packing 구조 -> 튜플

for entry in enumerate(['a','b','c']):
  print(entry)
```

```python
# 시작인덱스 변경

for i, idx in enumerate(['a','b','c'], start=1):
  print(i, idx)
```

컴프리헨션
  - 리스트, 튜플, 딕셔너리 안에 for문, if문ㅇ르 지정하여 생성하는 것
  - 형식
    1. 리스트
      - [식 for 변수 in 리스트]
      - list(식 for 변수 in 리스트)
      - list(식 for 변수 in 리스트 if 조건식)

    2. 튜플
      - (식 for 변수 in 리스트 if 조건식) : 사용X -> 제너레이터 표현식
      - tuple(식 for 변수 in 리스트)
      - tuple(식 for 변수 in 리스트 if 조건식)

```python
# 1. 반복문
l1 = []
for i in range(10):
  l1.append(i)
print(l1)
```

```python
# 2. range() 사용
l1 = list(range(10))
print(l1)
```

```python
# 3. 리스트컴프리헨션

l1 = [i for i in range(10)]
print(l1)

l2 = list(i for i in range(10))
print(l2)
```
